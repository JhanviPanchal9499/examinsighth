```python
# =========================
# INSTALL LIBRARIES
# =========================
!pip install flask pyngrok pandas openpyxl pypdf python-docx -q

import os
import re
import pandas as pd
from flask import Flask, request, jsonify, render_template_string, send_file
from pypdf import PdfReader
from docx import Document
from pyngrok import ngrok
import threading

# =========================
# CONFIG
# =========================
DB_FILE = "/content/Question_file.xlsx"   # Built-in question file path
SIMILARITY_THRESHOLD = 0.7

# Put your ngrok auth token here
NGROK_AUTH_TOKEN = "3Af61vDFxQRR8QuufuBDaV3DOQV_4AW45rRfmitz8nLe8Jj6M"

# =========================
# CHECK FILE EXISTS
# =========================
if not os.path.exists(DB_FILE):
    raise FileNotFoundError(
        "Question_file.xlsx not found in /content/. "
        "Upload it to Colab Files or copy it from Drive."
    )

# =========================
# FLASK APP
# =========================
app = Flask(__name__)

# =========================
# TEXT CLEANING + SIMILARITY
# =========================
def clean_text(text):
    text = str(text).lower().strip()
    text = re.sub(r'[^a-zA-Z0-9 ]', '', text)
    text = re.sub(r'\s+', ' ', text)
    return text

def similarity(q1, q2):
    q1 = clean_text(q1)
    q2 = clean_text(q2)

    w1 = set(q1.split())
    w2 = set(q2.split())

    if not w1:
        return 0

    return len(w1.intersection(w2)) / len(w1)

# =========================
# EXTRACT EXAM DETAILS
# =========================
def extract_semester(text):
    m = re.search(r'Sem[- ]?(\d+)', str(text), re.IGNORECASE)
    return m.group(1) if m else ""

def extract_exam_type(text):
    t = str(text).lower()
    if "summer" in t:
        return "Summer"
    if "winter" in t:
        return "Winter"
    return ""

def extract_year(text):
    m = re.search(r'20\d{2}', str(text))
    return m.group(0) if m else ""

# =========================
# LOAD EXCEL DATABASE
# =========================
def load_db():
    if not os.path.exists(DB_FILE):
        return []

    sheets = pd.read_excel(DB_FILE, sheet_name=None)
    data = []

    for subject_name, df in sheets.items():
        df = df.fillna("")

        for _, row in df.iterrows():
            exam_text = str(row.get("Exam", ""))

            data.append({
                "Serial No": row.get("Serial No", ""),
                "Subject": row.get("Subject", subject_name),
                "Exam": exam_text,
                "Section": row.get("Section", ""),
                "Question No": row.get("Question No", ""),
                "Question": str(row.get("Question", "")),
                "Marks": row.get("Marks", ""),
                "Semester": extract_semester(exam_text),
                "ExamType": extract_exam_type(exam_text),
                "Year": extract_year(exam_text)
            })

    return data

# =========================
# PDF / DOCX QUESTION EXTRACTION
# =========================
def extract_pdf(file):
    reader = PdfReader(file)
    text = ""

    for page in reader.pages:
        extracted = page.extract_text()
        if extracted:
            text += extracted + "\n"

    lines = [line.strip() for line in text.split("\n") if line.strip()]
    questions = []

    for line in lines:
        if re.match(r'^(Q\.?\s*\d+[)\]: -]*|[0-9]+\.)', line, re.IGNORECASE):
            line = re.sub(r'^(Q\.?\s*\d+[)\]: -]*|[0-9]+\.)\s*', '', line).strip()
            if line:
                questions.append(line)

    return questions

def extract_docx(file):
    doc = Document(file)
    questions = []

    for p in doc.paragraphs:
        line = p.text.strip()
        if re.match(r'^(Q\.?\s*\d+[)\]: -]*|[0-9]+\.)', line, re.IGNORECASE):
            line = re.sub(r'^(Q\.?\s*\d+[)\]: -]*|[0-9]+\.)\s*', '', line).strip()
            if line:
                questions.append(line)

    return questions

# =========================
# HTML
# =========================
HTML = """
<!DOCTYPE html>
<html>
<head>
    <meta charset="UTF-8">
    <title>ExamInsight Dashboard</title>
    <style>
        body {
            font-family: Arial, sans-serif;
            background: #f4f7fb;
            margin: 0;
            padding: 0;
        }
        .container {
            width: 90%;
            max-width: 1150px;
            margin: 30px auto;
            background: white;
            padding: 25px;
            border-radius: 14px;
            box-shadow: 0 4px 20px rgba(0,0,0,0.08);
        }
        h1 {
            text-align: center;
            color: #1f3c88;
            margin-bottom: 10px;
        }
        .stats {
            display: flex;
            gap: 20px;
            margin: 20px 0 30px 0;
            flex-wrap: wrap;
        }
        .card {
            flex: 1;
            min-width: 220px;
            background: #eef4ff;
            padding: 20px;
            border-radius: 12px;
            text-align: center;
            font-size: 18px;
            font-weight: bold;
            color: #1f3c88;
        }
        .section {
            margin-top: 25px;
            padding: 20px;
            background: #fafafa;
            border-radius: 12px;
        }
        .section h2 {
            margin-top: 0;
            color: #333;
        }
        input, select, button {
            padding: 10px;
            margin: 8px 5px 8px 0;
            border-radius: 8px;
            border: 1px solid #ccc;
            font-size: 14px;
        }
        button {
            background: #1f3c88;
            color: white;
            border: none;
            cursor: pointer;
        }
        button:hover {
            background: #162d66;
        }
        table {
            width: 100%;
            border-collapse: collapse;
            margin-top: 15px;
            background: white;
        }
        table, th, td {
            border: 1px solid #ddd;
        }
        th, td {
            padding: 10px;
            text-align: left;
            font-size: 14px;
            vertical-align: top;
        }
        th {
            background: #1f3c88;
            color: white;
        }
        .msg {
            margin-top: 10px;
            color: green;
            font-weight: bold;
        }
    </style>
</head>
<body>
<div class="container">
    <h1>ExamInsight Dashboard</h1>

    <div class="stats">
        <div class="card">Subjects: <span id="subjects">0</span></div>
        <div class="card">Questions: <span id="questions">0</span></div>
    </div>

    <div class="section">
        <h2>Search Single Question</h2>
        <input type="text" id="singleQuestion" placeholder="Enter question" style="width:70%;">
        <button onclick="searchQuestion()">Search</button>
        <div id="searchResult"></div>
    </div>

    <div class="section">
        <h2>Check Multiple Questions from PDF/DOCX</h2>
        <input type="file" id="multiFile" accept=".pdf,.docx">
        <button onclick="checkMultiple()">Check</button>
        <div id="multiResult"></div>
    </div>

    <div class="section">
        <h2>Insert Questions into Subject Sheet</h2>
        <input type="file" id="insertFile" accept=".pdf,.docx">
        <select id="subjectSelect"></select>
        <button onclick="insertQuestions()">Insert</button>
        <div class="msg" id="insertMsg"></div>
    </div>

    <div class="section">
        <h2>Download Updated Question File</h2>
        <a href="/download"><button>Download Excel File</button></a>
    </div>
</div>

<script>
async function loadStats() {
    const res = await fetch("/stats");
    const data = await res.json();
    document.getElementById("subjects").innerText = data.subjects;
    document.getElementById("questions").innerText = data.questions;
}

async function loadSubjects() {
    const res = await fetch("/subjects");
    const subjects = await res.json();
    const select = document.getElementById("subjectSelect");
    select.innerHTML = "";
    subjects.forEach(sub => {
        const option = document.createElement("option");
        option.value = sub;
        option.textContent = sub;
        select.appendChild(option);
    });
}

function makeTable(data) {
    let html = "<table><tr><th>Question</th><th>Status</th><th>Subject</th><th>Exam</th><th>Section</th><th>Question No</th><th>Semester</th><th>Year</th><th>Marks</th><th>Similarity %</th></tr>";
    data.forEach(item => {
        html += `<tr>
            <td>${item.question}</td>
            <td>${item.status}</td>
            <td>${item.subject}</td>
            <td>${item.exam}</td>
            <td>${item.section}</td>
            <td>${item.question_no}</td>
            <td>${item.semester}</td>
            <td>${item.year}</td>
            <td>${item.marks}</td>
            <td>${item.similarity}</td>
        </tr>`;
    });
    html += "</table>";
    return html;
}

async function searchQuestion() {
    const q = document.getElementById("singleQuestion").value;
    const res = await fetch("/search", {
        method: "POST",
        headers: {"Content-Type": "application/json"},
        body: JSON.stringify({q: q})
    });
    const data = await res.json();
    document.getElementById("searchResult").innerHTML = makeTable(data);
}

async function checkMultiple() {
    const fileInput = document.getElementById("multiFile");
    if (!fileInput.files.length) {
        alert("Please select a file");
        return;
    }

    const formData = new FormData();
    formData.append("file", fileInput.files[0]);

    const res = await fetch("/multi", {
        method: "POST",
        body: formData
    });

    const data = await res.json();
    document.getElementById("multiResult").innerHTML = makeTable(data);
}

async function insertQuestions() {
    const fileInput = document.getElementById("insertFile");
    const subject = document.getElementById("subjectSelect").value;

    if (!fileInput.files.length) {
        alert("Please select a file");
        return;
    }

    const formData = new FormData();
    formData.append("file", fileInput.files[0]);
    formData.append("subject", subject);

    const res = await fetch("/insert", {
        method: "POST",
        body: formData
    });

    const data = await res.json();
    document.getElementById("insertMsg").innerText = data.msg;
    loadStats();
}

loadStats();
loadSubjects();
</script>
</body>
</html>
"""

# =========================
# ROUTES
# =========================
@app.route("/")
def home():
    return render_template_string(HTML)

@app.route("/stats")
def stats():
    db = load_db()
    subject_count = len(set(item["Subject"] for item in db))
    question_count = len(db)
    return jsonify({"subjects": subject_count, "questions": question_count})

@app.route("/subjects")
def get_subjects():
    sheets = pd.read_excel(DB_FILE, sheet_name=None)
    return jsonify(list(sheets.keys()))

@app.route("/search", methods=["POST"])
def search_question():
    q = request.json.get("q", "").strip()
    db = load_db()

    results = []
    for item in db:
        score = similarity(q, item["Question"])
        if score >= SIMILARITY_THRESHOLD:
            results.append({
                "question": item["Question"],
                "status": "Found",
                "subject": item["Subject"],
                "exam": item["Exam"],
                "section": item["Section"],
                "question_no": item["Question No"],
                "semester": item["Semester"],
                "year": item["Year"],
                "marks": item["Marks"],
                "similarity": round(score * 100, 2)
            })

    if not results:
        results.append({
            "question": q,
            "status": "Not Found",
            "subject": "",
            "exam": "",
            "section": "",
            "question_no": "",
            "semester": "",
            "year": "",
            "marks": "",
            "similarity": 0
        })

    return jsonify(results)

@app.route("/multi", methods=["POST"])
def multi_question_check():
    file = request.files["file"]

    if file.filename.lower().endswith(".pdf"):
        questions = extract_pdf(file)
    elif file.filename.lower().endswith(".docx"):
        questions = extract_docx(file)
    else:
        return jsonify({"error": "Only PDF and DOCX files are supported"}), 400

    db = load_db()
    output = []

    for q in questions:
        best_score = 0
        best_match = None

        for item in db:
            score = similarity(q, item["Question"])
            if score > best_score:
                best_score = score
                best_match = item

        if best_match and best_score >= SIMILARITY_THRESHOLD:
            output.append({
                "question": q,
                "status": "Found",
                "subject": best_match["Subject"],
                "exam": best_match["Exam"],
                "section": best_match["Section"],
                "question_no": best_match["Question No"],
                "semester": best_match["Semester"],
                "year": best_match["Year"],
                "marks": best_match["Marks"],
                "similarity": round(best_score * 100, 2)
            })
        else:
            output.append({
                "question": q,
                "status": "Not Found",
                "subject": "",
                "exam": "",
                "section": "",
                "question_no": "",
                "semester": "",
                "year": "",
                "marks": "",
                "similarity": round(best_score * 100, 2)
            })

    return jsonify(output)

@app.route("/insert", methods=["POST"])
def insert_questions():
    file = request.files["file"]
    subject = request.form["subject"].strip()

    if file.filename.lower().endswith(".pdf"):
        questions = extract_pdf(file)
    elif file.filename.lower().endswith(".docx"):
        questions = extract_docx(file)
    else:
        return jsonify({"msg": "Only PDF and DOCX files are supported"}), 400

    sheets = pd.read_excel(DB_FILE, sheet_name=None)

    if subject not in sheets:
        sheets[subject] = pd.DataFrame(columns=[
            "Serial No", "Subject", "Exam", "Section", "Question No", "Question", "Marks"
        ])

    df = sheets[subject].fillna("")
    inserted = 0

    existing_questions = df["Question"].astype(str).tolist() if "Question" in df.columns else []

    next_serial = 1
    if "Serial No" in df.columns and len(df) > 0:
        numeric_serials = pd.to_numeric(df["Serial No"], errors="coerce").dropna()
        if not numeric_serials.empty:
            next_serial = int(numeric_serials.max()) + 1

    for q in questions:
        duplicate = False
        for old_q in existing_questions:
            if similarity(q, old_q) >= SIMILARITY_THRESHOLD:
                duplicate = True
                break

        if not duplicate:
            new_row = {
                "Serial No": next_serial,
                "Subject": subject,
                "Exam": "Unknown",
                "Section": "Unknown",
                "Question No": "Unknown",
                "Question": q,
                "Marks": "Unknown"
            }
            df.loc[len(df)] = new_row
            existing_questions.append(q)
            next_serial += 1
            inserted += 1

    sheets[subject] = df

    with pd.ExcelWriter(DB_FILE, engine="openpyxl") as writer:
        for sheet_name, sheet_df in sheets.items():
            sheet_df.to_excel(writer, sheet_name=sheet_name, index=False)

    return jsonify({"msg": f"{inserted} Questions Inserted Successfully"})

@app.route("/download")
def download_file():
    return send_file(DB_FILE, as_attachment=True)

# =========================
# RUN FLASK IN BACKGROUND
# =========================
def run_app():
    app.run(host="0.0.0.0", port=5000, debug=False, use_reloader=False)

thread = threading.Thread(target=run_app)
thread.start()

# =========================
# START NGROK
# =========================
ngrok.set_auth_token(NGROK_AUTH_TOKEN)

# Kill old tunnels if rerun
try:
    ngrok.kill()
except:
    pass

public_url = ngrok.connect(5000)
print("Dashboard URL:", public_url)
```
